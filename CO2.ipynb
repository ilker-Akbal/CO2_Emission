```python
X=df.drop("CO2_Emissions",axis=1)
y=df["CO2_Emissions"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=9)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
train_rmse = []
test_rmse = []

for d in range(1,10):
    
    poly_conv = PolynomialFeatures(degree=d, include_bias=False)
    poly_feat = poly_conv.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(poly_feat, y, test_size=0.3, random_state=99)
    
    model = LinearRegression(fit_intercept=True)
    model.fit(X_train,y_train)
    
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    
    
    train_RMSE = np.sqrt(mean_squared_error(y_train,train_pred))
    
    test_RMSE = np.sqrt(mean_squared_error(y_test,test_pred))

    train_rmse.append(train_RMSE)
    test_rmse.append(test_RMSE)
```

```python
plt.plot(range(1,6),train_rmse[:5],label='TRAIN')
plt.plot(range(1,6),test_rmse[:5],label='TEST')
plt.xlabel("Polinom Derecesi")
plt.ylabel("RMSE")
plt.legend()
```

```python
model = PolynomialFeatures(degree=3, include_bias=False)
```

```python
poly_feat = model.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_feat, y, test_size=0.3, random_state=99)
```

```python
model_poly = LinearRegression(fit_intercept=True)
```

```python
model_poly.fit(X_train, y_train)
```

```python
pred_poly = model_poly.predict(X_test)
```

```python
print("R^2 :       ", round(r2_score(y_test,pred_poly),2))
print("MAE :       ", round(mean_absolute_error(y_test,pred_poly),2))
print("RMSE :      ", round(np.sqrt(mean_squared_error(y_test,pred_poly)),2))
```
